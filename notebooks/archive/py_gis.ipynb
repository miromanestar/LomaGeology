```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd

df = pd.read_excel('../data/hawaii-5-11-22.xls', sheet_name='Complete')
```

```python
data = df[[
    'Latitude (Avg)',
    'Longitude (Avg)',
    # Major Elements
    'FeO*',
    "SIO2(WT%)", 
    "AL2O3(WT%)",
    "TIO2(WT%)",
    "CAO(WT%)", 
    "MGO(WT%)", 
    "K2O(WT%)", 
    "NA2O(WT%)",
    # Minor Elements
    # "P2O5(WT%)",
    # "MNO(WT%)", 
]]
data = data.dropna()

data.head()

print('Number of samples:', len(data.index))
```

```python
# Later, let's try to color the points by their kmeans cluster

hawaii = gpd.read_file('../shapefiles/hawaii')
fig, ax = plt.subplots(figsize=(100, 100))

hawaii.plot(color='lightgrey', ax=ax)
data.plot(x='Longitude (Avg)', y='Latitude (Avg)', kind='scatter', ax=ax, alpha=0.5)
```
